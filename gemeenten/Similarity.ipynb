```python
df = pd.read_csv('TK2017_revised.csv', sep=',' ,encoding= 'latin1')
df = df[['gemeente','Partij']]


df2 = pd.read_csv('2017_all.csv',sep=',',index_col=0)
df2 = df2.rename(columns={"Regio's": "gemeente"})
df2 = df2[np.isfinite(df2['Bevolking/Bevolkingssamenstelling op 1 januari/Totale bevolking (aantal)'])]

df3 = pd.merge(df,df2, on='gemeente')

df3 = df3[df3.columns[pd.Series(df3.columns).str.startswith('Inkomen')] | df3.columns[pd.Series(df3.columns).str.startswith('gemeente')]]



#columns = list(df3.columns[3:])

#columns = list(df3.columns[pd.Series(df3.columns).str.startswith('Bevolking')])


#df4 = df3.transpose()
#df4.columns = df4.iloc[1]
#df4.drop(df4.index[[0,1,2]],inplace=True)



#VVD = pd.DataFrame(df4.VVD)
 
#CDA = pd.DataFrame(df4.CDA)

#SP = pd.DataFrame(df4.CDA)

#SGP = pd.DataFrame(df4.CDA)

#PVDA = pd.DataFrame(df4.CDA) 

#PVV = pd.DataFrame(df4.CDA)


dd = pd.DataFrame(df3.pivot_table(columns='gemeente'))

corr_matrix = df3.corr().abs()





#the matrix is symmetric so we need to extract upper triangle matrix without diagonal (k = 1)
os = (corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool)).stack().sort_values(ascending=False))
#first element of os series is the pair with the bigest correlation

z = pd.DataFrame(os)

z
```

```python
from sklearn.metrics import classification_report
similarity = pd.read_csv('similarity_2017.csv')

similarity.columns = ['gemeente','gemeente2','correlatie']



final = pd.merge(similarity, df)

final = pd.merge(final, df, how='left', left_on=['gemeente2'], right_on=['gemeente'])


final = final[final.correlatie > 0.95].dropna()


a = final.Partij_x.values
b = final.Partij_y.values

final.groupby('Partij_y')['Partij_x'].value_counts()

print(classification_report(a, b))


```
